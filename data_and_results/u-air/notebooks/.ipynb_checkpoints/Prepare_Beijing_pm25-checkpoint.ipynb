### Analysis on which interpolation is most appropriate

```python
# print('Station ID', 'Number of time-stamps for which Weather info is available', sep=' '*5)
count_df = pd.DataFrame(index=data_df.station_id.unique(), columns=['count'])
for station in data_df.station_id.unique():
    count_df.loc[station, 'count'] = data_df[data_df.station_id==station]['temperature'].dropna().shape[0]
count_df = count_df.sort_values('count', ascending=False);
```

```python
perfect_stations = count_df.iloc[:24].index
```

```python
import GPy
```

#### Linear interpolation

```python
res_df = pd.DataFrame(np.zeros((3,4)), index=['temperature', 'pressure', 'humidity'], columns=['linear', 'quadratic','cubic','gp'])
for v_i, var in enumerate(['temperature', 'pressure', 'humidity']):
    for method in ['linear', 'quadratic','cubic','gp']:
        interpol_df = data_df[data_df.station_id.isin(perfect_stations)].set_index('station_id')[[var]]
        tmp_scaler = StandardScaler()
        interpol_df[interpol_df.columns] = tmp_scaler.fit_transform(interpol_df)
        s_rmse = []
        for s_i, station in enumerate(perfect_stations):
            tmp_df = interpol_df.loc[station]
            splitter = KFold(n_splits=5, shuffle=True, random_state=s_i)
            diffs = []
            for tt_i, (trn_i, tst_i) in enumerate(splitter.split(tmp_df.index)):
                tmp_df_trn = tmp_df.copy().reset_index(drop=True)
#                 tmp_df_trn += np.random.rand(tmp_df_trn.shape[0])*10**-5
                tmp_df_trn.iloc[tst_i] = None
                if s_i==0 and tt_i==0:
                    plt.figure(figsize=(15,4))
                    plt.plot(range(len(tmp_df)), tmp_df.values, 'o-', label='Ground Truth')
#                 print(tmp_df.values)
                if method == 'gp':
                    k = GPy.kern.RBF(1)
                    a_scaler = StandardScaler()
                    model = GPy.models.GPRegression(a_scaler.fit_transform(trn_i.reshape(-1,1)), tmp_df_trn.dropna().values, k)
                    model.optimize()
                    pred = model.predict(a_scaler.transform(tst_i.reshape(-1,1)))[0]
                    tmp_df_trn.loc[tst_i] = pred
                else:
                    tmp_df_trn = tmp_df_trn.interpolate(method=method, limit_direction='both')
                    tmp_df_trn = tmp_df_trn.interpolate('ffill').interpolate('bfill')
                if s_i==0 and tt_i==0:
                    plt.plot(tmp_df_trn.index, tmp_df_trn.values, '-', label='filled')
                    plt.title(method)
                    plt.xlabel('time-stamp')
                    plt.ylabel('temperature')
                    plt.legend()
                diffs.extend((np.abs(tmp_scaler.inverse_transform(tmp_df_trn.values[tst_i]) - tmp_scaler.inverse_transform(tmp_df.values[tst_i]))).tolist())
            s_rmse.append(np.mean(diffs))
#             print(var, method, station, 'RMSE:', s_rmse[-1])
        print(var, method, 'RMSE:', np.mean(s_rmse))
        res_df.loc[var, method] = np.mean(s_rmse)
```

```python
### Fill data
final_dfs = []
ipols = ['linear', 'quadratic']

for ipol in ipols:
    df_list = []
    for station in data_df.station_id.unique():
        tmp_df = data_df[data_df.station_id==station]
        tmp_df[['temperature','pressure','humidity','wind']] = tmp_df[['temperature','pressure','humidity','wind']].interpolate(ipol, limit_direction='both')
        tmp_df = tmp_df.interpolate('ffill').interpolate('bfill')
        df_list.append(tmp_df)
    
    final_df = pd.concat(df_list).sort_index()
    assert final_df.shape == final_df.dropna().shape # Checking presense of NA
    final_dfs.append(final_df)

linear_df, quadratic_df = final_dfs
linear_df = linear_df.sort_values(['time', 'station_id'])
quadratic_df = quadratic_df.sort_values(['time', 'station_id'])
linear_df.head(3)
```

```python
all_stations = linear_df.station_id.unique()
splitter = KFold(n_splits=6, random_state=42, shuffle=True)
names = ['linear/']#, 'quadratic/']
path = '../production/pm25_beijing_best36/'

Xcols = ['longitude', 'latitude', 'temperature', 'pressure', 'humidity', 'wind']
Xcols += ['weather_1.0', 'weather_2.0', 'weather_3.0', 'weather_5.0']
ycol = 'PM25_AQI_value'

init = time()
for df_i, df in enumerate([linear_df]):
    df = pd.get_dummies(df, columns=['weather'])
    for tv_i, (train_idx, test_idx) in enumerate(splitter.split(all_stations)):
        
        train_stations = all_stations[train_idx]
        test_stations = all_stations[test_idx]
        
        train_df = df[df.station_id.isin(train_stations)]
        test_df = df[df.station_id.isin(test_stations)]
        
        xscaler = StandardScaler()
        yscaler = StandardScaler()
        xscaler.fit(train_df[Xcols])
        yscaler.fit(train_df[[ycol]])
        
        cov_df = pd.crosstab(train_df.index, train_df.station_id, yscaler.transform(train_df[[ycol]])
                              , aggfunc = lambda x:x )
        
        for end in ['/train/X/', '/train/y', '/test/X/', '/test/y', '/scaler/']:
                if not os.path.exists(path+names[df_i]+'data'+str(len(Xcols))+'/fold_'+str(tv_i)+end):
                    os.makedirs(path+names[df_i]+'data'+str(len(Xcols))+'/fold_'+str(tv_i)+end)
        
        pd.to_pickle(sorted(train_stations), path+names[df_i]+'data'+str(len(Xcols))+'/fold_'+str(tv_i)+'/train_stations.pickle')
        pd.to_pickle(sorted(test_stations), path+names[df_i]+'data'+str(len(Xcols))+'/fold_'+str(tv_i)+'/test_stations.pickle')

        for t_i, ts in enumerate(df.index.unique()):
            clear_output(True)
            if t_i<24:
                emp_cov = cov_df.iloc[:24].cov().values
            else:
                emp_cov = cov_df.iloc[t_i-24:t_i].cov().values
            print(emp_cov.shape)
            np.savez_compressed(path+names[df_i]+'data'+str(len(Xcols))+'/fold_'+str(tv_i)+'/train/X/'+str(t_i).zfill(5)+'emp_cov.npz', emp_cov)
            
            train_x = xscaler.transform(train_df.loc[ts][Xcols])
            test_x = xscaler.transform(test_df.loc[ts][Xcols])
            
            train_y = yscaler.transform(train_df.loc[ts][[ycol]])
            test_y = test_df.loc[ts][[ycol]].values.reshape(-1,1)
            
            print(f'{df_i=} {tv_i=} {t_i=} len of train={train_x.shape}, len of test={test_x.shape}')
            
            np.savez_compressed(path+names[df_i]+'data'+str(len(Xcols))+'/fold_'+str(tv_i)+'/train/X/'+str(t_i).zfill(5)+'.npz', train_x)
            np.savez_compressed(path+names[df_i]+'data'+str(len(Xcols))+'/fold_'+str(tv_i)+'/test/X/'+str(t_i).zfill(5)+'.npz', test_x)
            
            np.savez_compressed(path+names[df_i]+'data'+str(len(Xcols))+'/fold_'+str(tv_i)+'/train/y/'+str(t_i).zfill(5)+'.npz', train_y)
            np.savez_compressed(path+names[df_i]+'data'+str(len(Xcols))+'/fold_'+str(tv_i)+'/test/y/'+str(t_i).zfill(5)+'.npz', test_y)
            
            pd.to_pickle(yscaler, path+names[df_i]+'data'+str(len(Xcols))+'/fold_'+str(tv_i)+'/scaler/'+str(t_i).zfill(5)+'.pickle')
print('\nDataset ready in',(time()-init)/60,'minutes')
```

```python
def Cluster_GP(trn_x, trn_y, tst_x):
    np.random.seed(0)
#     print(trn_x.shape)
    
    model = GPy.models.GPRegression(trn_x, trn_y-trn_y.mean(), GPy.kern.RBF(trn_x.shape[1], ARD=True), normalizer=False)
    model.kern.lengthscale = trn_x.max(axis=0) - trn_x.min(axis=0)
    model.kern.variance = np.var(trn_y)
    model['Gaussian.noise'] = np.var(trn_y)/10
#     print(model)
#     sys.exit()
#     print(model, model.kern.lengthscale)
    model.optimize_restarts(10, robust=True, verbose=False, parallel=False)
#     model.plot()
#     print(model, model.kern.lengthscale)
#     sys.exit()
#     X1, X2 = np.meshgrid(np.linspace(trn_x[:,0].min()-0.1, trn_x[:,0].max()+0.1, 100), 
#                      np.linspace(trn_x[:,1].min()-0.1, trn_x[:,1].max()+0.1, 100))
#     X_test = np.array([(x1, x2) for x1, x2 in zip(X1.ravel(), X2.ravel())])
    
#     pred_y = model.predict(X_test)[0] + trn_y.mean()
#     plt.contourf(X1, X2, pred_y.reshape(100,100), alpha=0.3)
#     plt.colorbar()
#     sys.exit()
    return model.predict(tst_x)[0] + trn_y.mean()
```

```python
init = time()
all_stations = linear_df.station_id.unique()
splitter = KFold(n_splits=6, random_state=42, shuffle=True)
df = pd.get_dummies(linear_df.copy(), columns=['weather'])
res_df_test = pd.DataFrame(index=range(277), columns=all_stations)
res_df_pred = pd.DataFrame(index=range(277), columns=all_stations)

Xcols = ['longitude', 'latitude']#, 'temperature', 'pressure', 'humidity', 'wind']
ycol = 'PM25_AQI_value'
# Xcols = Xcols + ['weather_1.0', 'weather_2.0', 'weather_3.0', 'weather_5.0']

for tv_i, (train_idx, test_idx) in enumerate(splitter.split(all_stations)):
#     plt.scatter(all_locs[:, 0], all_locs[:, 1], c=c)
    for c_i in set(c[test_idx]):
        c_trn_idx = train_idx[c[train_idx] == c_i]
        c_tst_idx = test_idx[c[test_idx] == c_i]
#         plt.scatter(all_locs[c_trn_idx, 0], all_locs[c_trn_idx, 1], marker='d', c='r')
#         plt.scatter(all_locs[c_tst_idx, 0], all_locs[c_tst_idx, 1], marker='d', c='g')
#         sys.exit()
        train_stations = all_stations[c_trn_idx]
        test_stations = all_stations[c_tst_idx]
        
        train_df = df[df.station_id.isin(train_stations)]
        test_df = df[df.station_id.isin(test_stations)]
        
#         xscaler = StandardScaler()
#         yscaler = StandardScaler()
#         xscaler.fit(train_df[Xcols])
#         yscaler.fit(train_df[[ycol]])
        
        for t_i, ts in enumerate(df.index.unique()):
            clear_output(True)
            
#             train_x = xscaler.transform()
            train_x = train_df.loc[ts][Xcols].values.reshape(-1,len(Xcols))
#             test_x = xscaler.transform(test_df.loc[ts][Xcols].values.reshape(-1,len(Xcols)))
            test_x = test_df.loc[ts][Xcols].values.reshape(-1,len(Xcols))
            
#             train_y = yscaler.transform(train_df.loc[ts][[ycol]].values.reshape(-1,1))
            train_y = train_df.loc[ts][[ycol]].values.reshape(-1,1)
            test_y = test_df.loc[ts][[ycol]].values.reshape(-1,1)
            
            pred_y = Cluster_GP(train_x, train_y, test_x)
            
            res_df_test.loc[t_i, test_stations] = test_y.ravel()
            res_df_pred.loc[t_i, test_stations] = pred_y.ravel()
            
            print(f'{tv_i=} {c_i=} {t_i=} len of train={train_x.shape}, len of test={test_x.shape}')
print(time()-init)
```

```python
def load_results(path, folds, f_ids, n_test, model_name): # n_test = number of test stations
    pred_y = np.zeros((len(folds), len(f_ids), n_test))#*np.nan
    test_y = np.zeros((len(folds), len(f_ids), n_test))#*np.nan
    for fold_id, fold in enumerate(folds):
        for file_id, file in enumerate(f_ids):
            try:
                pred_y[fold_id, file_id, :] = np.load(path+'results/'+model_name+'/fold_'+fold+'/'+file+'.npz')['arr_0'].ravel()
                test_y[fold_id, file_id, :] = np.load(path+'data/fold_'+fold+'/test/y/'+file+'.npz')['arr_0'].ravel()
            except:
                print(model_name,fold_id,file,'not found')
    return pred_y, test_y

res_path = '../production/pm25_beijing_best36/linear/'
el_res_test = pd.DataFrame(index=range(277), columns=all_stations)
el_res_pred = pd.DataFrame(index=range(277), columns=all_stations)

load_el_pred, load_el_test = load_results(res_path, [str(i) for i in range(6)], 
                                          [str(i).zfill(5) for i in range(277)], 6, 'gp_rbf_gpy')
for tv_i, (train_idx, test_idx) in enumerate(splitter.split(all_stations)):
    el_res_test.iloc[:, test_idx] = load_el_test[tv_i]
    el_res_pred.iloc[:, test_idx] = load_el_pred[tv_i]
```

# Generate train, test for ADAIN

```python
all_stations = linear_df.station_id.unique()
splitter = KFold(n_splits=6, random_state=42, shuffle=True)
names = ['linear/', 'quadratic/']
path = '../production/pm25_beijing_best36/'
Xcols = ['longitude', 'latitude', 'temperature', 'pressure', 'humidity', 'wind']
met_cols = ['temperature', 'pressure', 'humidity', 'wind']
ycol = ['PM25_AQI_value']
ws = 24

init = time()
for df_i, df in enumerate([linear_df, quadratic_df]):
    for tv_i, (train_idx, test_idx) in enumerate(splitter.split(all_stations)):
        train_stations = all_stations[train_idx]
        test_stations = all_stations[test_idx]
        
        train_df = df[df.station_id.isin(train_stations)]
        test_df = df[df.station_id.isin(test_stations)]
        
        xscaler = StandardScaler()
        train_df[Xcols] = xscaler.fit_transform(train_df[Xcols])
        test_df[Xcols] = xscaler.transform(test_df[Xcols])
        
        yscaler = StandardScaler()
        train_df[ycol] = yscaler.fit_transform(train_df[ycol])
        
        for end in ['/train/adain', '/test/adain']:
            if not os.path.exists(path+names[df_i]+'data/fold_'+str(tv_i)+end):
                os.makedirs(path+names[df_i]+'data/fold_'+str(tv_i)+end)
        
        ## For training
        
        stn_dist_list = []
        stn_met_aq_list = []
        local_met_list = []
        local_aqi_list = []
        tst_stn_dist_list = []
        tst_stn_met_aq_list = []
        tst_local_met_list = []
        tst_local_aqi_list = []
        
        t_idx = df.index.unique()
        for t_i, ts in enumerate(t_idx[:-ws+1]):
            clear_output(True)
            
            print(f'{df_i=} {tv_i=} {t_i=}')
            
            f_ts = t_idx[t_i]
            l_ts = t_idx[t_i+ws-1]
            stn_df = train_df.loc[f_ts:l_ts][met_cols+ycol+['station_id']]
            tst_stn_df = test_df.loc[f_ts:l_ts][met_cols+ycol+['station_id']]
            ## For test
            tst_dist = np.nan*np.zeros((1,2,len(train_stations),len(test_stations)))
            tst_loc_met = np.nan*np.zeros((1,ws,len(met_cols),len(test_stations)))
            for tst_idx, test_station in enumerate(test_stations):
                local_loc = station_df.loc[test_station][['longitude','latitude']].values.reshape(-1,2)
                station_loc = station_df[station_df.index.isin(train_stations)][['longitude','latitude']].values
                tst_dist[0,:,:,tst_idx] = (station_loc-local_loc).T
                
                # Saving local met sample
                tst_loc_met[0,:,:,tst_idx] = tst_stn_df[tst_stn_df.station_id==test_station][met_cols].values
                
            # Saving dist samples
            tst_stn_dist_list.append(tst_dist)
            
            # Saving met_aq sample
            stn_met_aq = stn_df[stn_df.station_id.isin(train_stations)].drop(columns=['station_id'])
            ans = np.concatenate([i[:,:,None] for i in np.split(stn_met_aq.values, len(train_stations))], axis=2)
            tst_stn_met_aq_list.append(ans[None,:,:,:])
#             print(ans[None,:,:,:].shape)
            
            # Saving loc_met sample
            tst_local_met_list.append(tst_loc_met)
#             print(tst_loc_met.shape)
            
            # Saving aqi for loca
            tst_local_aqi_list.append(tst_stn_df.loc[l_ts][ycol].values.T)
            
            # For train
            for local_station in train_stations:
                tmp_stations = train_stations.copy().tolist()
                tmp_stations.remove(local_station)
                
                local_loc = station_df.loc[local_station][['longitude','latitude']].values.reshape(-1,2)
                station_loc = station_df[station_df.index.isin(tmp_stations)][['longitude','latitude']].values
                
                # Saving dist samples
                stn_dist_list.append((station_loc-local_loc).T[None,:,:])
                
                # Saving train_met_aq sample
                stn_met_aq = stn_df[stn_df.station_id.isin(tmp_stations)].drop(columns=['station_id'])
                ans = np.concatenate([i[:,:,None] for i in np.split(stn_met_aq.values, len(tmp_stations))], axis=2)
                stn_met_aq_list.append(ans[None,:,:,:])
                
                # Saving local_met sample
                local_met_list.append(stn_df[stn_df.station_id==local_station][met_cols].values[None,:,:])
                
                # Saving test aqi
                local_aqi_list.append(stn_df[stn_df.station_id==local_station].loc[l_ts][ycol].values)
                
#         print(np.concatenate(stn_dist_list).shape)
#         print(np.concatenate(stn_met_aq_list).shape)
#         print(np.concatenate(local_met_list).shape)
#         print(np.concatenate(local_aqi_list).shape)
#         sys.exit()
        np.savez_compressed(path+names[df_i]+'data/fold_'+str(tv_i)+'/train/adain/trn_dst.npz', np.concatenate(stn_dist_list))
        np.savez_compressed(path+names[df_i]+'data/fold_'+str(tv_i)+'/train/adain/trn_metaq.npz', np.concatenate(stn_met_aq_list))
        np.savez_compressed(path+names[df_i]+'data/fold_'+str(tv_i)+'/train/adain/tst_met.npz', np.concatenate(local_met_list))
        np.savez_compressed(path+names[df_i]+'data/fold_'+str(tv_i)+'/train/adain/tst_aqi.npz', np.concatenate(local_aqi_list))
        
        np.savez_compressed(path+names[df_i]+'data/fold_'+str(tv_i)+'/test/adain/trn_dst.npz', np.concatenate(tst_stn_dist_list))
        np.savez_compressed(path+names[df_i]+'data/fold_'+str(tv_i)+'/test/adain/trn_metaq.npz', np.concatenate(tst_stn_met_aq_list))
        np.savez_compressed(path+names[df_i]+'data/fold_'+str(tv_i)+'/test/adain/tst_met.npz', np.concatenate(tst_local_met_list))
        np.savez_compressed(path+names[df_i]+'data/fold_'+str(tv_i)+'/test/adain/tst_aqi.npz', np.concatenate(tst_local_aqi_list))
        
        if not os.path.exists(path+names[df_i]+'data/fold_'+str(tv_i)+'/scaler/adain/'):
            os.makedirs(path+names[df_i]+'data/fold_'+str(tv_i)+'/scaler/adain/')
        
        pd.to_pickle(yscaler, path+names[df_i]+'data/fold_'+str(tv_i)+'/scaler/adain/yscaler.pickle')
        
print('\nDataset ready in',(time()-init)/60,'minutes')
```
