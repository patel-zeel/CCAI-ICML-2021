```python
import pandas as pd
import numpy as np
from time import time
import matplotlib.pyplot as plt
from matplotlib import rc
from scipy.spatial.distance import cdist
import GPy
from polire.interpolate import Idw

rc('font', size=16)
```

### Loading AQ and ERA5 weather data

```python
aq_version = '0.1.0'
era5_version = '1.1.0' # old versions: 0.1.0 - single levels data
aq_df = pd.read_pickle('../processed_data/aq_v'+aq_version+'.pickle')
era5_df = pd.read_pickle('../processed_data/weather_v'+era5_version+'.pickle')
```

### Plotting close AQ and ERA5 locations

```python
era5_locations = np.unique(np.vstack([era5_df.longitude.values, era5_df.latitude.values]), axis=1).T
era5_locations.shape
```

```python
tmp_df = aq_df.drop_duplicates(subset=['location'])

# Finding closest ERA5 locations
aq_long_lat = []
era_long_lat = []
for long, lat in zip(tmp_df.longitude, tmp_df.latitude):
    closest_idx = np.argmin((cdist(np.array([long, lat]).reshape(1,2), era5_locations)).ravel())
    aq_long_lat.append([long, lat])
    era_long_lat.append([era5_locations[closest_idx][0], era5_locations[closest_idx][1]])
    
aq_long_lat = np.array(aq_long_lat)
era_long_lat = np.array(era_long_lat)
    
# Plotting
plt.figure(figsize=(12,5))
plt.scatter(aq_long_lat[:,0], aq_long_lat[:,1], c='k', label='AQ stations')
plt.scatter(era_long_lat[:,0], era_long_lat[:,1], c='g', marker='d', s=200, label='Closest ERA5 locations')
for aq, era in zip(aq_long_lat, era_long_lat):    
    plt.plot([aq[0], era[0]], [aq[1], era[1]])
plt.legend(bbox_to_anchor=(1,1));
```

#### ERA5 single levels data will not work here because 37 AQ stations share only 4 locations of ERA5 data

#### With ERA5 land data, 37 AQ stations share 13 locations of ERA5 data. As per above figure, selecting closest observation will not work, so, we will interpolate weather variables in the region and accordingly select the observations per AQI stations.

### Interpolating ERA5 data

Bilinear interpolation on ERA5 data

Longitude range: 76.9 - 77.4    
Latitude range: 28.4 - 28.9

#### Selecting data in delhi band

```python
era5_short_df = era5_df[(era5_df.longitude>=76.9) & (era5_df.longitude<77.45)]
era5_short_df = era5_short_df[(era5_short_df.latitude>=28.4) & (era5_short_df.latitude<28.95)]
era5_short_df = era5_short_df.set_index('time').sort_index()

era5_short_df.shape
era5_short_df.head(2)
```

#### Interpolating over AQ station locations

```python
tmp_df = aq_df.drop_duplicates(subset=['location'])
interpol_loc = tmp_df.location.values
interpol_lonlat = tmp_df[['longitude', 'latitude']].values
interpol_loc.shape, interpol_lonlat.shape
```

```python
imp_cols = era5_short_df.columns[2:]
def interpolate(X, y, X_new):
    model = Idw(exponent=1)
    model.fit(X, y)
    return model.predict(X_new)

train_df = pd.DataFrame(index=era5_short_df.index.unique(), columns=imp_cols, dtype='object')

# Prepare train data
init = time()
print(train_df.shape)
for t in train_df.index:
    for col in train_df.columns:
        X, y = era5_short_df.loc[t, ['longitude', 'latitude']].values, era5_short_df.loc[t, col].values.reshape(-1,1)
        train_df.loc[t, col] = [{'X':X,'y':y}]
print('Preparing finished in',round((time()-init)/60, 2),'minutes')
```

```python
interpol_df = train_df.copy()

# Interpolating data
init = time()
for col in interpol_df:
    print(col,end=" ")
    interpol_df[col] = interpol_df[col].apply(lambda x: interpolate(x[0]['X'], x[0]['y'], interpol_lonlat))
print('Interpolation finished in',round((time()-init)/60, 2),'minutes')
```

### Choosing Aug data

There is diversity in missing data along stations. So, one may need to discard stations based on time-frame chosen.

```python
aq_df_aug = aq_df_new.loc['2019-08-01':'2019-08-31']
```

### Visualize missing data

```python
base_length = len(pd.date_range('2019-08-01','2019-08-31 23:59:59', freq='1H'))
missing_data_df = pd.DataFrame(index=aq_df_aug.location.unique(), columns=['frac'])

for loc in aq_df_aug.location.unique():
    tmp_df = aq_df_aug[aq_df_aug.location==loc].resample('1H').agg({'pm25':np.mean})
    missing_data_df.loc[loc, 'frac'] = tmp_df.dropna(subset=['pm25']).shape[0]/base_length
    
miss_df = missing_data_df.sort_values('frac', ascending=False).reset_index()
miss_df
```
